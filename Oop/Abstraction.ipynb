```python
#Abstract Class
from abc import ABC, abstractmethod  # You have to import these for abstraction
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

#Note you can't create and instance of Abstract class
```

```python
# Concrete Class you have to use all the function define in Shape

class Rectangle(Shape):
    def __init__(self,length: int, breadth: int):
        self.length=length
        self.breadth=breadth
    def area(self)->int:
        print(f"Area of rectangle is {self.length * self.breadth}")
```

```python
rectangle=Rectangle(4,2)
rectangle.area() # it will throw error since the concrete class has not implemented all the function of Abstract Class
```

```python
# Concrete Class you have to use all the function define in Shape

class Rectangle(Shape):
    def __init__(self,length: int, breadth: int):
        self.length=length
        self.breadth=breadth
    def area(self)->int:
        print(f"Area of rectangle is {self.length * self.breadth}")
    def perimeter(self)->int:
        print(f"Perimeter of rectangle is {2*(self.length + self.breadth)}")
```

```python
rectangle=Rectangle(4,2)
rectangle.area() 
```
